```python
from sklearn.neighbors import KNeighborsClassifier
from  sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv('mt_cars.csv')

base.head()
```

```python
X = base[['mpg', 'hp']].values

y = base['cyl'].values
```

```python
knn = KNeighborsClassifier(n_neighbors=3)

model = knn.fit(X, y)
```

```python
predictions = model.predict(X)

print(predictions)
```

```python
accuracy = accuracy_score(y, predictions)

precision = precision_score(y, predictions, average='weighted') # Com peso

recall = recall_score(y, predictions, average='weighted')

f1 = f1_score(y, predictions, average='weighted')

print(f'Acurácia: {accuracy}, precisão: {precision}, recall: {recall}, F1: {f1}')
```

```python
cm = confusion_matrix(y, predictions)

print(cm)
```

```python
new_data = np.array([[19.3, 105]])

prediction_it = model.predict(new_data)

print(prediction_it)
```

```python
distancies, indexes = model.kneighbors(new_data)

print(distancies)

print(indexes)
```
