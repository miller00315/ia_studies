```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
data = h2o.import_file('insurance.csv')
```

```python
data
```

```python
data = data.drop('C1')
```

```python
train, test = data .split_frame(ratios=[0.7], seed= 0)
```

```python
model_auto_ml = H2OAutoML(max_runtime_secs=180, sort_metric='AUTO')

model_auto_ml.train(y='Accident', training_frame=train)

```

```python
ranking = model_auto_ml.leaderboard

ranking = ranking.as_data_frame()

ranking
```

```python
prever = model_auto_ml.leader.predict(test)

```
